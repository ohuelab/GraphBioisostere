```python
# df_marariaの全てのSMILESを構造式として可視化
from rdkit import Chem
from rdkit.Chem import Draw
from IPython.display import display

# SMILESから分子オブジェクトを作成
mols = []
legends = []

for idx, row in df_mararia.iterrows():
    smiles = row['SMILES']
    
    # SMILESが文字列型でない、または空の場合はスキップ
    if not isinstance(smiles, str) or not smiles or pd.isna(smiles):
        print(f"Warning: Invalid or missing SMILES at index {idx}: {smiles}")
        continue
    
    mol = Chem.MolFromSmiles(smiles)
    
    if mol is not None:
        mols.append(mol)
        # MMV Numberをラベルとして使用
        mmv_number = row.get('MMV Number', f'Index {idx}')
        legends.append(str(mmv_number))
    else:
        print(f"Warning: Cannot parse SMILES at index {idx}: {smiles}")

print(f"\nTotal valid molecules: {len(mols)}")

# グリッド表示で可視化
if len(mols) > 0:
    img = Draw.MolsToGridImage(
        mols, 
        molsPerRow=8, 
        subImgSize=(300, 300), 
        legends=legends
    )
    display(img)
else:
    print("No valid molecules to display")
```

```python
# 可視化できた化合物かつSMILESが"FC"で始まる化合物のみを抽出
valid_indices = []

for idx, row in df_mararia.iterrows():
    smiles = row['SMILES']
    
    # SMILESが文字列型でない、または空の場合はスキップ
    if not isinstance(smiles, str) or not smiles or pd.isna(smiles):
        continue
    
    # RDKitで解析できるかチェック
    mol = Chem.MolFromSmiles(smiles)
    
    # 有効な分子でかつSMILESが"FC"で始まる場合
    if mol is not None and smiles.startswith('FC'):
        valid_indices.append(idx)

# 該当する行のみを抽出
df_mararia_flt = df_mararia.loc[valid_indices].copy()
df_mararia_flt.reset_index(drop=True, inplace=True)

print(f"Total FC-starting compounds: {len(df_mararia_flt)}")
df_mararia_flt
```

```python
# 必要なモジュールとパスの設定
import sys
import torch
import torch.nn as nn
from torch_geometric.data import Batch

# bioiso/pro_GNNディレクトリをパスに追加
sys.path.insert(0, '/home/8/uf02678/gsbsmasunaga/bioiso/pro_GNN')

from utils.loader import smiles_to_data, load_encoder
from config import Args

# モデルの定義（training_cls_ddp_3.pyから）
def make_head(in_dim: int, hidden: int, out_dim: int = 2) -> nn.Sequential:
    """2-layer MLP ヘッドを生成"""
    return nn.Sequential(
        nn.Linear(in_dim, hidden),
        nn.ReLU(),
        nn.Linear(hidden, hidden // 2),
        nn.ReLU(),
        nn.Linear(hidden // 2, out_dim)
    )

class GraphDiffClassifier(nn.Module):
    def __init__(self, encoder: nn.Module, embedding_dim: int,
                 loss_type: str, hidden_dim: int = 64, out_dim: int = 2, merge_method: str = "diff"):
        super().__init__()
        assert loss_type in {"pair", "pair_bi", "pair_bi_sym"}
        self.loss_type = loss_type
        self.encoder   = encoder
        self.merge_method = merge_method

        self.mlp = make_head(embedding_dim, hidden_dim, out_dim)

    def forward(self, data_a, data_b):
        fa = self.encoder(data_a)
        fb = self.encoder(data_b)

        if self.merge_method == "diff":
            diff_feature = torch.abs(fa - fb)
            pred = self.mlp(diff_feature)
        elif self.merge_method == "product":
            product_feature = fa * fb
            pred = self.mlp(product_feature)
        else:
            raise ValueError(f"Unknown merge_method: {self.merge_method}")

        return pred

print("Modules loaded successfully")
```

```python
# cv0~4の全モデルを読み込み
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Argsの設定（training時と同じ設定を使用）
args = Args()
args.embedding_dim = 64
args.hidden_dim = 64
args.loss_type = "pair"
args.model_type = "diff"

# cv0~4の各モデルを読み込む
models = []
base_path = "/home/8/uf02678/gsbsmasunaga/bioiso/pro_GNN/results/results_consistentsmiles"

for cv_idx in range(5):
    print(f"\nLoading model from cv{cv_idx}...")
    
    # エンコーダーとモデルの構築
    encoder = load_encoder(args).to(device)
    model = GraphDiffClassifier(
        encoder,
        embedding_dim=args.embedding_dim,
        loss_type=args.loss_type,
        hidden_dim=args.hidden_dim,
        out_dim=2,
        merge_method=args.model_type
    ).to(device)
    
    # 学習済みモデルの読み込み
    model_path = f"{base_path}/cv{cv_idx}/pair-diff/model_best.pt"
    state_dict = torch.load(model_path, map_location=device, weights_only=False)
    model.load_state_dict(state_dict)
    model.eval()
    
    models.append(model)
    print(f"  Model loaded from: {model_path}")

print(f"\nTotal models loaded: {len(models)}")
```

```python
# df_pairsの各SMILESペアをグラフデータに変換して、全モデルで推論
import numpy as np
from tqdm import tqdm

def predict_bioisostere_score_ensemble(models, smiles1, smiles2, device):
    """
    2つのSMILESから生物学的等価体スコアを予測（アンサンブル）
    """
    try:
        # SMILESをグラフデータに変換（1回だけ）
        data1 = smiles_to_data(smiles1)
        data2 = smiles_to_data(smiles2)
        
        # バッチ化してデバイスに転送
        batch1 = Batch.from_data_list([data1]).to(device)
        batch2 = Batch.from_data_list([data2]).to(device)
        
        # 各モデルで推論
        scores = []
        with torch.no_grad():
            for model in models:
                pred = model(batch1, batch2)
                # Softmaxで確率に変換
                prob = torch.softmax(pred, dim=1)
                # クラス1（生物学的等価体である）の確率を取得
                score = prob[0, 1].item()
                scores.append(score)
        
        return scores
    except Exception as e:
        print(f"Error processing SMILES pair: {e}")
        return [None] * len(models)

# df_pairsの各ペアに対してスコアを計算
all_scores = {f'cv{i}_score': [] for i in range(5)}
all_scores['mean_score'] = []
all_scores['std_score'] = []

for idx, row in tqdm(df_pairs.iterrows(), total=len(df_pairs), desc="Predicting scores"):
    scores = predict_bioisostere_score_ensemble(
        models,
        row['Ref_SMILES'],
        row['Cand_SMILES'],
        device
    )
    
    # 各cvのスコアを記録
    for cv_idx, score in enumerate(scores):
        all_scores[f'cv{cv_idx}_score'].append(score)
    
    # 平均と標準偏差を計算
    if None not in scores:
        all_scores['mean_score'].append(np.mean(scores))
        all_scores['std_score'].append(np.std(scores))
    else:
        all_scores['mean_score'].append(None)
        all_scores['std_score'].append(None)

# スコアをデータフレームに追加
for key, values in all_scores.items():
    df_pairs[key] = values

# 平均スコアでソート
df_pairs['Bioisostere_Score'] = df_pairs['mean_score']  # 互換性のため
df_pairs.sort_values(by='mean_score', ascending=False, inplace=True)

print(f"\nPrediction completed for {len(df_pairs)} pairs using {len(models)} models")
print(f"Columns added: {list(all_scores.keys())}")

```

```python
# Ref_IC50とCand_IC50の前処理：>が含まれる場合は100に変更
def clean_ic50(ic50_value):
    if isinstance(ic50_value, str):
        # >が含まれている場合は100に設定
        if '>' in ic50_value:
            return 100.0
        # >がない場合は数値に変換
        ic50_value = ic50_value.strip()
    try:
        return float(ic50_value)
    except (ValueError, TypeError):
        return np.nan

df_pairs['Ref_IC50'] = df_pairs['Ref_IC50'].apply(clean_ic50)
df_pairs['Cand_IC50'] = df_pairs['Cand_IC50'].apply(clean_ic50)

print("IC50 values cleaned:")
print(f"Ref_IC50 - Min: {df_pairs['Ref_IC50'].min():.2f}, Max: {df_pairs['Ref_IC50'].max():.2f}")
print(f"Cand_IC50 - Min: {df_pairs['Cand_IC50'].min():.2f}, Max: {df_pairs['Cand_IC50'].max():.2f}")
```

```python
# IC50の対数差分を計算（活性の改善度）
df_pairs['ΔpIC50'] = np.log10(df_pairs['Ref_IC50']) - np.log10(df_pairs['Cand_IC50'])
```

## 順位相関の評価

```python
# Bioisostere_ScoreとΔpIC50の順位相関を評価
from scipy.stats import spearmanr, kendalltau
import matplotlib.pyplot as plt

# NaN値を除外
df_eval = df_pairs_subset.dropna(subset=['Bioisostere_Score', 'ΔpIC50'])

print("=" * 60)
print("順位相関の評価: Bioisostere_Score vs ΔpIC50")
print("=" * 60)

# 1. スピアマンの順位相関係数
spearman_corr, spearman_pval = spearmanr(df_eval['Bioisostere_Score'], df_eval['ΔpIC50'])
print(f"\n【スピアマンの順位相関係数】")
print(f"  相関係数 (ρ): {spearman_corr:.4f}")
print(f"  p値: {spearman_pval:.4e}")
print(f"  解釈: ", end="")
if abs(spearman_corr) > 0.7:
    print("強い相関")
elif abs(spearman_corr) > 0.4:
    print("中程度の相関")
elif abs(spearman_corr) > 0.2:
    print("弱い相関")
else:
    print("ほとんど相関なし")

# 2. ケンドールの順位相関係数（τ）
kendall_corr, kendall_pval = kendalltau(df_eval['Bioisostere_Score'], df_eval['ΔpIC50'])
print(f"\n【ケンドールの順位相関係数】")
print(f"  相関係数 (τ): {kendall_corr:.4f}")
print(f"  p値: {kendall_pval:.4e}")

# 3. 順位の比較
df_eval['Score_Rank'] = df_eval['Bioisostere_Score'].rank(ascending=False, method='average')
df_eval['DeltaIC50_Rank'] = df_eval['ΔpIC50'].rank(ascending=False, method='average')
df_eval['Rank_Diff'] = abs(df_eval['Score_Rank'] - df_eval['DeltaIC50_Rank'])

print(f"\n【順位の一致度】")
print(f"  平均順位差: {df_eval['Rank_Diff'].mean():.2f}")
print(f"  中央値順位差: {df_eval['Rank_Diff'].median():.2f}")
print(f"  最大順位差: {df_eval['Rank_Diff'].max():.0f}")
print(f"  順位差の標準偏差: {df_eval['Rank_Diff'].std():.2f}")

# 完全一致の割合
exact_match = (df_eval['Rank_Diff'] == 0).sum()
within_1 = (df_eval['Rank_Diff'] <= 1).sum()
within_2 = (df_eval['Rank_Diff'] <= 2).sum()

print(f"\n【順位の近似度】")
print(f"  完全一致: {exact_match}/{len(df_eval)} ({exact_match/len(df_eval)*100:.1f}%)")
print(f"  ±1位以内: {within_1}/{len(df_eval)} ({within_1/len(df_eval)*100:.1f}%)")
print(f"  ±2位以内: {within_2}/{len(df_eval)} ({within_2/len(df_eval)*100:.1f}%)")

# 4. トップ5, トップ10の一致度
top5_score = set(df_eval.nlargest(5, 'Bioisostere_Score').index)
top5_delta = set(df_eval.nlargest(5, 'ΔpIC50').index)
top10_score = set(df_eval.nlargest(10, 'Bioisostere_Score').index)
top10_delta = set(df_eval.nlargest(10, 'ΔpIC50').index)

print(f"\n【トップランク化合物の一致度】")
print(f"  トップ5の一致: {len(top5_score & top5_delta)}/5 ({len(top5_score & top5_delta)/5*100:.1f}%)")
print(f"  トップ10の一致: {len(top10_score & top10_delta)}/10 ({len(top10_score & top10_delta)/10*100:.1f}%)")

print("\n" + "=" * 60)
```

```python
# 順位相関の可視化
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# 左: 順位の散布図
ax1 = axes[0]
ax1.scatter(df_eval['Score_Rank'], df_eval['DeltaIC50_Rank'], alpha=0.7, s=100)
ax1.plot([1, len(df_eval)], [1, len(df_eval)], 'r--', linewidth=2, label='Perfect Agreement')
ax1.set_xlabel('Bioisostere Score Rank', fontsize=12)
ax1.set_ylabel('ΔpIC50 Rank', fontsize=12)
ax1.set_title(f'Rank Comparison\n(Spearman ρ={spearman_corr:.3f})', fontsize=14)
ax1.legend()
ax1.grid(True, alpha=0.3)
ax1.invert_xaxis()  # スコアが高い方を左に
ax1.invert_yaxis()  # ΔpIC50が高い方を上に

# 右: 順位差のヒストグラム
ax2 = axes[1]
ax2.hist(df_eval['Rank_Diff'], bins=range(0, int(df_eval['Rank_Diff'].max())+2), 
         edgecolor='black', alpha=0.7, color='steelblue')
ax2.axvline(df_eval['Rank_Diff'].mean(), color='r', linestyle='--', 
            linewidth=2, label=f'Mean: {df_eval["Rank_Diff"].mean():.2f}')
ax2.set_xlabel('Rank Difference (|Score Rank - ΔpIC50 Rank|)', fontsize=12)
ax2.set_ylabel('Frequency', fontsize=12)
ax2.set_title('Distribution of Rank Differences', fontsize=14)
ax2.legend()
ax2.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()

# トップ10のリストを表示
print("\n【Bioisostere Score トップ10】")
top10_by_score = df_eval.nlargest(10, 'Bioisostere_Score')[['Bioisostere_Score', 'ΔpIC50', 'Score_Rank', 'DeltaIC50_Rank', 'Rank_Diff']]
print(top10_by_score.to_string())

print("\n【ΔpIC50 トップ10】")
top10_by_delta = df_eval.nlargest(10, 'ΔpIC50')[['Bioisostere_Score', 'ΔpIC50', 'Score_Rank', 'DeltaIC50_Rank', 'Rank_Diff']]
print(top10_by_delta.to_string())
```
